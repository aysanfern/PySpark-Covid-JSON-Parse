##Loading and parsing JSON

```
from pyspark.sql.functions import col
from itertools import islice
from pyspark.sql.functions import explode, col


df = spark.read.json("/FileStore/shared_uploads/aysanfernandes@gmail.com/covid.json",multiLine=True)

#using explode is necessary top be able to retrieve each patient as it's own seperate row rather than being in the same level of the list
dfnew = df.withColumn("parsed",explode(df["covid"]))


dfnew.rdd.take(5)

```

```
#creating for loop to parse each element in the row to a new column
for i in range(0,23):
  dx = dfnew.withColumn(str(i),dfnew['parsed'][i])
  dfnew=dx


#dropping unnecessary columns
dx = dx.drop('covid')
dx = dx.drop('parsed')


##retrieving the first row which contains all of the headers and saving it as a variable so it can be attached to the dataframe later
row = dx.where(col("0")=='#')
first= row.first()


#removing the first row from the dataframe by enumerating the dataframe and removing the row with index == 0 
adapted = dx.rdd.mapPartitionsWithIndex(
    lambda idx, it: islice(it, 1, None) if idx == 0 else it 
)
dx = adapted.toDF()



#renaming all of the columns to have the same name as the first rows
for i in dx.columns:
  dx = dx.withColumnRenamed(i,first[i])

#renaming the column with '#' to 'id' for clarity
dx = dx.withColumnRenamed('#','id')
dx.show()
```

##Analysis: Indexing and creating a data pipeline

```
from pyspark.ml.linalg import Vectors
from pyspark.ml.feature import StringIndexer
from pyspark.ml import Pipeline

#creating an indexer for loop to iterate through each column
indexers = [StringIndexer(inputCol=column, outputCol=column+"_index").fit(dx) for column in list(set(dx.columns)-set(['id'])-set(['age'])) ]


#creating a pipeline so that it will streamline this process when fitting the data
pipeline = Pipeline(stages=indexers)
DF6 = pipeline.fit(dx).transform(dx)

DF6.show(2)
```

```
import pandas as pd

#working out the sum of the columns this tells us how many of the dataset had the respective symptom e.g 100 means 100 people from the dataset had said symptom
list = DF6.groupby().sum().collect()


#converting this list into a dictionary
dictionary = list[0].asDict()




#creating a dataframe from this dictionary
dataframe= pd.DataFrame.from_dict(dictionary,orient='index')


#filtering the dataframe for values where symptoms are greater than 110 meaning we can find the most frequent symptoms
top_symptoms=dataframe[dataframe[0]>110]

#resetting index so that the dataframe can have it's own index and we can use the existing index for our x axis values for our visualisation stage
top_symptoms=top_symptoms.reset_index()


#sorting the dataframe in descending order for visualisation purposes
top_symptoms= top_symptoms.sort_values(by=[0] ,ascending=False)

top_symptoms.head()
```

#Visualisation

```
top_symptoms.plot(kind='barh', x='index', y=0, colormap='winter_r')
```

```
top_symptoms.plot(kind='bar', x='index', y=0, colormap='winter_r')
```
